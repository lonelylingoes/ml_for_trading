## 优化器示例——求极值点

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as spo


def f(X):
    Y = (X - 1.5)**2 + 0.5
    print("X = {}, Y = {}".format(X, Y))
    return Y

def test_run():
    Xguess = 2.0
    min_result = spo.minimize(f, Xguess, method='SLSQP', options={'disp': True})
    print("Minima found at:")
    print("X = {}, Y = {}".format(min_result.x, min_result.fun))

    # 画图
    xplot = np.linspace(0.5, 2.5, 21)
    yplot = f(xplot)
    plt.plot(xplot, yplot)
    plt.plot(min_result.x, min_result.fun, 'ro')
    plt.title('Minima of an objective function')
    plt.show()
    
    
if __name__ == "__main__":
    test_run()

```

## 带目标函数的优化器实例——拟合直线

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as spo


def error(line, data):
    '''
    计算数据与直线直接的均方差
    参数：
        line: 直线的斜率和截距,tuple
        data: 真实数据,numpy array
    '''
    # 计算y轴的均方差之和
    err = np.sum((data[:, 1] - (line[0]*data[:, 0] + line[1]))**2)
    return err



def fit_line(data, error_func):
    '''
    参数：
        data:数据点， 2D numpy arra,shape of [n, 2]
        error_func: 误差函数
    返回拟合直线的斜率和截距
    '''
    # 设置初始值
    l = np.float32([0, np.mean(data[:, 1])])
    
    # 画出最原始的猜测
    x_ends = np.float32([-5, 5])
    plt.plot(x_ends, l[0] *x_ends + l[1], 'm--', linewidth=2.0, label="Inital guess")
    
    # 调用优化器求解
    result = spo.minimize(error_func, l, args=(data, ), method='SLSQP', options={'disp':True})
    return result.x


def test_fun():
    # 定义初始线
    l_orig = np.float32([4, 12])
    print("Orinal line: C0 = {}, C1={}".format(l_orig[0], l_orig[1]))
    x_orig = np.linspace(0, 10, 21)
    y_orig = l_orig[0] * x_orig + l_orig[1]
    plt.plot(x_orig, y_orig, 'b--', linewidth = 2.0, label='Original line')
    
    # 在原始点上，生成带噪声的点
    noise_sigma = 3.0
    noise = np.random.normal(0, noise_sigma, y_orig.shape)
    data = np.array([x_orig, y_orig+noise]).T
    plt.plot(data[:, 0], data[:, 1], 'go', label='Data points')
    
    # 拟合直线
    l_fit = fit_line(data, error)
    print("Fitted line: C0 = {}, C1={}".format(l_fit[0], l_fit[1]))
    plt.plot(data[:,0], l_fit[0] * data[:, 0]+ l_fit[1], "r--", linewidth=2.0, label="Fit line")

    # 添加 legend 并画图
    plt.legend(loc='best')
    plt.show()

    
if __name__ == "__main__":
    test_fun()
```

## 带目标函数的优化器实例——拟合多项式曲线

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as spo


def error(C, data):
    '''
    计算数据与直线直接的均方差
    参数：
        C: 多项式系数,numpy array
        data: 真实数据,numpy array
    '''
    # 计算y轴的均方差之和
    err = np.sum((data[:, 1] - np.polyval(C, data[:, 0])) ** 2)
    return err


def fit_poly(data, error_func, degree=3):
    '''
    参数：
        data:数据点， 2D numpy arra,shape of [n, 2]
        error_func: 误差函数
        degree: 维度值
    返回拟合直线的斜率和截距
    '''
    # 设置初始多项式模型系数
    c_guess = np.poly1d(np.ones(degree + 1, dtype=np.float32))

    # 画出最原始的猜测
    x = np.linspace(-5, 8, 21)
    plt.plot(x, np.polyval(c_guess, x), 'm--', linewidth=2.0, label="Inital guess")

    # 调用优化器求解
    result = spo.minimize(error_func, c_guess, args=(data,), method='SLSQP', options={'disp': True})
    return np.poly1d(result.x)


def test_fun():
    # 定义初始线
    c_orig = np.float32([1.5, -10, -5, 60, 50])
    print(np.poly1d(c_orig))
    
    x_orig = np.linspace(0, 10, 21)
    y_orig = np.polyval(np.poly1d(c_orig), x_orig)
    plt.plot(x_orig, y_orig, 'b--', linewidth=2.0, label='Original line')

    # 在原始点上，生成带噪声的点
    noise_sigma = 3.0
    noise = np.random.normal(0, noise_sigma, y_orig.shape)
    data = np.array([x_orig, y_orig + noise]).T
    plt.plot(data[:, 0], data[:, 1], 'go', label='Data points')

    # 拟合多项式曲线
    c_fit = fit_poly(data, error, degree=4)
    print(c_fit)
    plt.plot(data[:, 0], np.polyval(c_fit, data[:, 0]), "r--", linewidth=2.0, label="Fit line")

    # 添加 legend 并画图
    plt.legend(loc='best')
    plt.show()


if __name__ == "__main__":
    test_fun()
```
