## 画出均线

```python
import pandas as pd
import numpy as np
import tushare as ts
import matplotlib.pyplot as plt

# 将日期作为索引，并修改收盘价为股票名
hs300 = ts.get_k_data('hs300', start='2012-01-01', end='2012-05-31')
hs300 = hs300.set_index('date')
hs300 = hs300.rename(columns={'close':'hs300'})[['hs300']]

rolling_mean = hs300.rolling(window=20).mean()
ax = hs300.plot(title='mean and close', label='hs300')
rolling_mean.plot(label='mean', ax=ax)
# Add axis labels and legend
ax.set_xlabel("Date")
ax.set_ylabel("Price")
ax.legend(loc='upper left')
plt.show()
```

## 画出布林带

```python
rolling_mean = hs300.rolling(window=20).mean()
rolling_std =  hs300.rolling(window=20).std()
upper_band = rolling_mean - rolling_std
lower_band = rolling_mean + rolling_std
ax = hs300.plot(title='Bollinger Bands', label='hs300')
ax = rolling_mean.plot(label='mean', ax=ax)
ax = upper_band.plot(label='upper band', ax=ax)
ax = lower_band.plot(label='lower band', ax=ax)
# Add axis labels and legend
ax.set_xlabel("Date")
ax.set_ylabel("Price")
ax.legend(loc='upper left')
plt.show()
```

## 画日收益率(daily returns)

```python
daily = hs300.copy()
daily[1:] = daily[1:] / daily.values[:-1] - 1
daily.iloc[0] = 0
ax = daily.plot(title='daily returns')
ax.set_xlabel("Date")
ax.set_ylabel("returns")
plt.show()
```

## 累积收益率(cumulative returns)

其实跟归一化后的价格图走势一模一样的，只是数值上差了1。

```python
cumulative = hs300.copy()
cumulative = cumulative / cumulative.iloc[0] - 1
ax = cumulative.plot(title='cumulative returns')
ax.set_xlabel("Date")
ax.set_ylabel("returns")
plt.show()
```
